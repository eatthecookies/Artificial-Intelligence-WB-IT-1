```python
import matplotlib.pyplot as plt
import math
import numpy as np
def f(x):
    return (((1 + math.e ** (x ** 0.5) + math.cos(x**2)) ** 0.5) / (abs(1 - math.sin(x) ** 3)))


x = [f(i) for i in range(1, 11)]
x1 = x[0:len(x) // 2]

# точечный график для среза первой половины массива
plt.plot([i for i in range(5)], x1, 'bo')
plt.show()

# линейный  график для основного массива
plt.plot([i for i in range(10)], x)
plt.show()
```

```python
import numpy as np
import math
from scipy.integrate import simps 
from numpy import trapz 
import matplotlib.pyplot as plt

x = np.arange(0,10,0.1)
y = np.abs(np.cos(x * math.e ** (np.cos(x) + np.log(x + 1))))

plt.grid()
plt.plot(x, y, c = 'r')
plt.fill_between(x,y)

area = trapz(y)
print(area)
```
